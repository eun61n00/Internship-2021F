```python
def search_glowpick(brand_name):
    from selenium import webdriver
    from selenium.webdriver.common.by import By
    from selenium.webdriver.common.keys import Keys
    from selenium.webdriver.support.ui import WebDriverWait
    from selenium.webdriver.support import expected_conditions as EC
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(ChromeDriverManager().install(), options=chrome_options)
    wait = WebDriverWait(driver, 10)
    try:

        link = 'https://www.glowpick.com/searches/keywords'
        brand_name = brand_name['brand_name']
        driver.get(link)
        time.sleep(1)
        html = driver.page_source
        soup = BeautifulSoup(html,"html.parser")
        remove_button = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="default-layout"]/div/div[1]/span/div/div[2]/div[2]/button[1]')))
        remove_button.click()
        element_input = driver.find_element_by_class_name('header__input')
        element_input.send_keys(brand_name)
        time.sleep(3)
        html = driver.page_source
        soup = BeautifulSoup(html,"html.parser")
        search_brand_name = soup.select('.products__brand__name')
        #print(search_brand_name)
        serch_brand_name_h_e = []
        for sbn in search_brand_name:
            serch_brand_name_h_e.append(sbn.text)
    except:
        serch_brand_name_h_e=[]
        pass
    driver.close()
    return serch_brand_name_h_e

```
